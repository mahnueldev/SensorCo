```python
import requests
import json
import random
import time
from datetime import datetime

# 🔗 Power BI push URL
PUSH_URL = "https://api.powerbi.com/beta/a9531b3d-965b-4eb8-8c10-a8f55a606e94/datasets/1dc4610b-0210-498a-9f8d-5b7a55fe9d73/rows?experience=power-bi&key=xdVxddv%2FLoncPwglXgsQkhXv4%2Fpjs8qVwUOGORFDB5XGTWRxomQ%2BkrHtN3s7xLiEv6XRYah7JRFUwrL93ylqsw%3D%3D"

# Starting temperature
temperature = random.randint(50, 100)

print("🌡️ Sending live sensor data to Power BI... (Press CTRL+C to stop)\n")

# Continuous loop
try:
    i = 0
    while True:
        i += 1
        speed = random.randint(0, 120)
        direction = random.randint(0, 360)  # use integer degrees
        temperature = random.uniform(temperature, min(temperature + 5, 200))

        data = [{
            "speed": speed,
            "direction": direction,
            "temperature": temperature,
            "timestamp": datetime.now().isoformat()
        }]

        response = requests.post(PUSH_URL, data=json.dumps(data))
        print(f"[{i}] Sent → Speed: {speed} km/h | Dir: {direction}° | Temp: {temperature:.1f}°C | Status: {response.status_code}")

        # Adjust interval (2 sec recommended for live dashboard)
        time.sleep(2)

except KeyboardInterrupt:
    print("\n🛑 Stream stopped by user.")
```
